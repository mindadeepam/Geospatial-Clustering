```python

# int(a)
def func(x):
    try : 
        x = float(x)
    except :
        x = 0.0
    return x
```

```python
df["longitude"] = df.longitude.apply(func)
df["latitude"] = df.latitude.apply(func)

df = df.loc[(df.latitude!=0.000) & (df.longitude!=0.000)] 
df = df.loc[(df.longitude<=98) & (df.longitude>=60)]
df = df.loc[(df.latitude<=38) & (df.latitude>=8)]
```

```python
df = df.groupby(['student_id','question_id']).first().reset_index()
```
